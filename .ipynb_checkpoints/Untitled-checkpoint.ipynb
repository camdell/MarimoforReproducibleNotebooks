```python
from polars import read_csv, col

df = read_csv('raw_training_study.csv')
df.head()
```

```python
clean_df = (
    df.filter(
        col('hours', 'score').is_not_null(),
        col('score').is_between(0, 100)
    )
    .with_columns(
        col('score') / 100
    )
)

clean_df.head()
```
